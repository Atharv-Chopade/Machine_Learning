```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
X = df.drop('target' , axis = 1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.32
                                                       , random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion= 'gini', max_depth=5 ,random_state=32)
```

```python
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report , ConfusionMatrixDisplay
```

```python
plt.figure(figsize=(5,10))
ConfusionMatrixDisplay.from_predictions(y_test , y_pred)
plt.show()
```

```python
plt.figure(figsize = (5,10))
ConfusionMatrixDisplay.from_estimator(model , X_test , y_test)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(confusion_matrix(y_test , y_pred) , annot = True)
```

```python
print(classification_report(y_test , y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
lg = LogisticRegression()
lg.fit(X_train_scaled , y_train)
y_pred_lg = lg.predict(X_test_scaled)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train_scaled , y_train)
y_pred_knn = knn.predict(X_test_scaled)
```

```python
dt = DecisionTreeClassifier(criterion='gini', max_depth = 6)
dt.fit(X_train , y_train)
y_pred_dt = dt.predict(X_test)
```

```python
models = ['Logistic Regression' , 'KNN' , 'Decision Tree']
accuracy = [accuracy_score(y_test , y_pred_lg) , accuracy_score(y_test , y_pred_knn) , accuracy_score(y_test , y_pred_dt)]
```

```python
plt.figure(figsize = (10,5))
plt.bar(models , accuracy)
plt.title('Models Comparison')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.show()
```
